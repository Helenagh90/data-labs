# Challenge 1

#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Para mostrar gr√°ficas en JN en vez de lanzarlas fuera:
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
fig, (ax1) = plt.subplots()
ax1.plot(x, y, "--")
ax1.set_title("XY")

fig, (ax2) = plt.subplots()
ax2.plot(x, z, "--", color = "y")
ax2.set_title("XZ")

plt.show();
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True)
ax1.plot(x, y, "--", color = "green")
ax2.plot(x, z, "--", color = "red")
ax1.set_title("XY")
ax2.set_title("XZ")

plt.show();
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (15, 5))
ax1.plot(x, y, "--", color = "green")
ax2.plot(x, z, "--", color = "red")
ax1.set_title("XY")
ax2.set_title("XZ")

plt.show();
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (15,5))
ax1.plot(x, y, "--", color = "green")
ax2.plot(x, z, "--", color = "red")
ax1.set_title("XY")
ax2.set_title("XZ")

ax1.set_xlabel("X values")
ax1.set_ylabel("Y values")
ax2.set_xlabel("X values")
ax2.set_ylabel("Y values")

plt.show();
```

#### Plot both `y=x^2` and `y=exp(x)`¬†in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
x = np.arange(0,10)
y1 = x**2
y2 = np.exp(x)

plt.figure(figsize = (15,5))
plt.plot(x, y1, "--", color = "green")
plt.plot(x, y2, "--", color = "red")

#When you are calling your figure using matplotlib.pyplot, use plt.xscale() instead of plt.set_xscale()!!
plt.xscale("linear")
plt.yscale("log")
```
