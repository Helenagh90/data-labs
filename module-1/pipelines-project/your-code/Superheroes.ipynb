```python
import pandas as pd
```

```python
def get_csv():
    df = pd.read_csv("superheroes.csv", index_col = 0, encoding = "latin-1", na_values=["-"])
    return df

superheroes = get_csv()
```

```python
def rename_column(df, old_column, new_column):
    return df.rename(columns = {old_column: new_column}, inplace = True)

superherores = rename_column(superheroes, "name", "Name")

superheroes.head()
```

```python
def rename_column_values(df, column, old_value, new_value):
    return df[column].replace(to_replace = {old_value: new_value}, inplace=True)

rename_column_values(superheroes, "Gender", "Male", "M")
rename_column_values(superheroes, "Gender", "Female", "F")

superheroes.head()
```

```python
def capitalize_column_values(df, column):
    df[column] = df[column].str.capitalize()

capitalize_column_values(superheroes, "Alignment")
capitalize_column_values(superheroes, "Eye color")
capitalize_column_values(superheroes, "Skin color")

superheroes.head()
```

```python
def pounds_to_kilograms(df, column):
    for value in df[column]:
        if value > 0:
            df[column].replace(to_replace = {value: round(value/2.205, 2)}, inplace=True)
        else:
            df[column].replace(to_replace = {value: None}, inplace=True)
        
pounds_to_kilograms(superheroes, "Weight")

superheroes.head()
```

```python
def feet_to_meters(df, column):
    for value in df[column]:
        if value > 0:
            df[column].replace(to_replace = {value: round(value/3.281, 2)}, inplace=True)
        else:
            df[column].replace(to_replace = {value: None}, inplace=True)
        
feet_to_meters(superheroes, "Height")

superheroes.head()
```

```python
def change_columns_order(df, list_new_columns_order):
    return df[list_new_columns_order]

superheroes = change_columns_order(superheroes, ["Name", "Race", "Gender", "Height", "Weight", "Skin color", "Hair color", "Eye color", "Alignment", "Publisher"])

superheroes.head()
```

```python
def total_nulls(df):
    return df.isnull().sum()

total_nulls(superheroes)
```

```python
def delete_unuseful_columns(df):
    for column in df:
        if df[column].isnull().sum() > len(df.index)/2:
            del df[column]

delete_unuseful_columns(superheroes)
    
superheroes.head()
```

```python
def rename_nulls(df):
    df.fillna("-", inplace=True)
    
rename_nulls(superheroes)

superheroes.head()
```
