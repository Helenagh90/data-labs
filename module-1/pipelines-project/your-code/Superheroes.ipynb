```python
import pandas as pd
```

```python
def acquire():
    df = pd.read_csv("superheroes.csv", index_col = 0, encoding = "latin-1")
    return df

superheroes = acquire()
superheroes.head()
```

```python
def rename_columns(df, old_column, new_column):
    return df.rename(columns = {old_column: new_column}, inplace = True)

superherores = rename_columns(superheroes, "name", "Name")
superheroes.head()
```

```python
def delete_column_nulls(df, column):
    for data in list(df.column):
        if data < 0:
            data.replace(data, "-")
    return 
superheroes = delete_column_nulls(superheroes, "Height")
superheroes.head()
```
